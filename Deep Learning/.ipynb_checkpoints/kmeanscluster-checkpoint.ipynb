```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import cv2
from collections import Counter
```

```python
### BACKGROUND'I SÄ°LME
```

```python
img = cv2.imread("hmgoepprod.jpg")
img = cv2.resize(img,(400,500))
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img_gray,cmap="gray")
```

```python
img_gray
```

```python
img_gray[img_gray>=220] = 0
```

```python
plt.imshow(img_gray,cmap="gray")
```

```python
ret,thresh1 = cv2.threshold(img_gray,127,255,cv2.THRESH_BINARY)
```

```python
thresh1.shape
```

```python
plt.imshow(thresh1,cmap="gray")
```

```python
kernel1 = np.ones((7,7),dtype=np.float32)
kernel2 = np.ones((2,2),dtype=np.float32)
```

```python
thresh1 = cv2.dilate(thresh1,kernel1)
```

```python
plt.imshow(thresh1,cmap="gray")
```

```python
thresh1 = cv2.erode(thresh1,kernel2)
```

```python
plt.imshow(thresh1,cmap="gray")
```

```python
mask1 = thresh1
```

```python
mask1 = cv2.bitwise_or(img,img,mask=mask1)
```

```python
plt.imshow(mask1)
```

```python
mask1.shape
```

```python
for i in mask1:
    for j in i:
        if j[0] == 0 & j[1] == 0 & j[2] == 0 :
            j[0] = -3
            j[1] = -3
            j[2] = -3
```

```python
### PREPROCESSING
```

```python
flatten_img = mask1.reshape((mask1.shape[0]*mask1.shape[1],3))
```

```python
### BUILDING MODEL 
```

```python
number_of_colors = 3  ### CLUSTER SAYISI
```

```python
model = KMeans(n_clusters = number_of_colors)
labels = model.fit_predict(flatten_img)
```

```python
counts = Counter(labels)
```

```python
colors = model.cluster_centers_
```

```python
def RGB2HEX(color):
    return "#{:02x}{:02x}{:02x}".format(int(color[0]), int(color[1]), int(color[2]))
```

```python
colors
```

```python
counts
```

```python
ordered_colors = [colors[i] for i in counts.keys()]
```

```python
ordered_colors
```

```python
hex_colors = [RGB2HEX(ordered_colors[i]) for i in counts.keys()]
```

```python
plt.figure(figsize = (8, 6))
plt.pie(counts.values(), labels = hex_colors, colors = hex_colors)
```

```python
hex_colors
```

```python
ordered_counts = np.zeros((3))
a = 0
```

```python
for i in counts.values():
    ordered_counts[a] = i
    a += 1
```

```python
ordered_counts = np.sort(ordered_counts,axis=-1)
```

```python
ordered_counts
```

```python
plt.figure(figsize = (8, 6))
plt.pie(ordered_counts, labels = hex_colors, colors = hex_colors)
```
