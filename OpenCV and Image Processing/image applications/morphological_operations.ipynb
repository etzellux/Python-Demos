```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    blank_img = np.zeros((600,600))
    cv2.putText(blank_img,"ABCDE",(50,300),cv2.FONT_HERSHEY_SIMPLEX,5,(255,255,255),20)
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap="gray")
```

```python
img = load_img()
```

```python
display_img(img)
```

```python
###############################################
############ EROSION ##########################
##############################################
```

```python
kernel = np.ones((5,5),dtype=np.uint8)
```

```python
result = cv2.erode(img,kernel,iterations=3)
```

```python
display_img(result)
```

```python
##########################################
############# OPENING ###################
########################################
# OPENING = EROSION + DILATION 
```

```python
img = load_img()
```

```python
white_noise = np.random.randint(0,2,(600,600))*255
```

```python
noise_img = white_noise + img
```

```python
display_img(noise_img)
```

```python
opening = cv2.morphologyEx(noise_img,cv2.MORPH_OPEN,kernel)
```

```python
display_img(opening)
```

```python
########################################################
########### CLOSING ###################################
######################################################
```

```python
img = load_img()
```

```python
black_noise = np.random.randint(0,2,(600,600))*(-255)
```

```python
noise_img = black_noise + img
```

```python
noise_img[noise_img == -255] = 0
```

```python
display_img(noise_img)
```

```python
closing = cv2.morphologyEx(noise_img,cv2.MORPH_CLOSE,kernel,iterations=1)
```

```python
display_img(closing)
```

```python
##############################################
###### MORPH GRADIENT #######################
############################################
# DILATION -EROSION
```

```python
img = load_img()
```

```python
display_img(img)
```

```python
gradient = cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel,iterations=1)
```

```python
display_img(gradient)
```
