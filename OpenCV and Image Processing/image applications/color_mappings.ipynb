```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("original/Computer-Vision-with-Python/DATA/00-puppy.jpg")
```

```python
plt.imshow(img)
```

```python
# OpenCV default olarak grseli bgr kanalna gre at
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #bgr-rgb'ye dntrld
```

```python
plt.imshow(img)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_RGB2HSV) #kanal rgb'den hsv'ye dntrld
```

```python
plt.imshow(img)
```
