```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread("titanfall2.jpg")
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.imread("ealogo.png")
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
img2 = cv2.resize(img2,(300,300))
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
ret,img2gray = cv2.threshold(img2,20,255,cv2.THRESH_BINARY)
```

```python
img2gray.shape
```

```python
mask_inv = img2gray[:,:,0]
```

```python
plt.imshow(mask_inv,cmap="gray")
```

```python
roi = img1[0:300,0:300,:]
```

```python
plt.imshow(roi)
```

```python
roi2 = cv2.bitwise_or(img2gray,roi)
```

```python
plt.imshow(roi2)
```

```python
final_roi = cv2.bitwise_xor(roi2,img2)
```

```python
plt.imshow(final_roi)
```
