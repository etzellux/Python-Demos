```python
img1 = cv2.imread("original/Computer-Vision-with-Python/DATA/dog_backpack.png")
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.imread("original/Computer-Vision-with-Python/DATA/watermark_no_copy.png")
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```
