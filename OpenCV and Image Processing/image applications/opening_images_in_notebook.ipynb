```python
img_gray = cv2.imread("original/Computer-Vision-with-Python/DATA/00-puppy.jpg",cv2.IMREAD_GRAYSCALE)
```

```python
new_img = cv2.resize(fix_img,(1000,400))
```

```python
plt.imshow(new_img)
```

```python
new_img = cv2.resize(fix_img,(0,0),fix_img,w_ratio,h_ratio) #GÖRSELİ BELİRLİ ORANDA BİÇİMLENDİRİR
```
